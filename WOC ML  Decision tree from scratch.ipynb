# Decision Tree Classifier

```python
class Node():
    def __init__(self, feature_index=None, threshold=None, left=None, right=None, gain=None, value=None):
        self.feature_index = feature_index
        self.threshold = threshold
        self.left = left
        self.right = right
        self.gain = gain
        self.value = value
```

```python
class DecisionTreeClassifier():
    
    def __init__(self, m , n):
        self.root = None
        self.m = m
        self.n = n
    
    
    def entropy(self,target_col):
        elements,counts = np.unique(target_col,return_counts = True)
        E = 0
        for i in range(len(elements)):
            E+=(((-1)*counts[i]/np.sum(counts))*np.log2(counts[i]/np.sum(counts)))
        return E
    
    
    def information_gain(self,parent,left,right,mode="entropy"):
        weight_l = len(left) / len(parent)
        weight_r = len(right) / len(parent)
        gain = self.entropy(parent) - (weight_l*self.entropy(left) + weight_r*self.entropy(right))
        return gain

    
    
```
