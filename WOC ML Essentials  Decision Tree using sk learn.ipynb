* Label Encoding for features of dataset having string value

```python
from sklearn.preprocessing import LabelEncoder
```

```python
cols = ["Gender","purpose_of_travel","Type_of_Travel","Type_Of_Booking","overall_experience"]
```

```python
le = LabelEncoder()
for col in cols:
    df[col] = le.fit_transform(df[col])
```

```python
df.head()
```

* Preparing the data

```python
df1 = df.drop("id",axis=1) 
```

```python
df1.head()
```

```python
df2 = df1.drop('Unnamed: 0',axis=1)
```

```python
df2.head()
```

```python
X = df2.drop('overall_experience', axis=1)
Y = df2['overall_experience']
```

```python
X.head()
```

```python
Y.head()
```

* Splitting dataset for training and testing

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.20)
```

```python
X_train.head()
```

```python
Y_train.head()
```

```python
X_test.head()
```

```python
Y_test.head()
```

# Decision Tree Clasifier using sk learn

* Importing Algorithm

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier()
```

* Training Algorithm

```python
classifier.fit(X_train, Y_train)
```

* Prediction

```python
Y_pred = classifier.predict(X_test)
```

* Evaluation

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(Y_test, Y_pred))
print(classification_report(Y_test, Y_pred))
```
