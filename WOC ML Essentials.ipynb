* Applying KMeans algorithm from scratch

```python
X = data_train.iloc[:,:2].values                            #slicing the training dataset
```

```python
X.shape
```

```python
k=4                                                           #Declaring number of clusters=4
```

```python
import random                                                 #initializing the centeroids randomly from training set
ran_centro = random.sample(range(0, len(X)), 4)             
ran_centro                                                    #output shows indices of randomly chosen centeroids
```

```python
cent = []                                            #finding data points of initially assumed centeroids and stored in a list 
for i in ran_centro:
    cent.append(X[i])
```

```python
centro = np.array(cent)                                      #declaring a separate array for initially assumed centeroids
print(centro)
```

```python
plt.scatter(data_train['Skills'],data_train['Hardwork'],alpha=0.3)
plt.xlabel('Skills')
plt.scatter(centro[:,0],centro[:,1],color='black',marker='*')
plt.ylabel('Hardwork')
plt.show()                                                 #plotting training data set and initially randomly assumed centeroids
```

```python
def dist(a,b):                                            #Defining a function to calculate euclidean distance 
    return(sum((a - b)**2))**0.5
```

```python
def short(m, n):                            #Defining a function to assign cluster to data points on basis of closest centeroid
    S = []
    for i in n:
        K = []
        for j in m:
            K.append(dist(i, j))
        S.append(np.argmin(K))
    return S
```

```python
def closest(Y, X):                           
    Z = []
    B = pd.concat([pd.DataFrame(X),pd.DataFrame(Y,columns=['cluster'])],axis=1)
    for c in set(B['cluster']):
        L = B[B['cluster'] == c][B.columns[:-1]]
        P = L.mean(axis=0)
        Z.append(P)
    return Z
```

```python
for i in range(10):                                      #Number of iterations=10
    N = short(centro, X)
    centro = closest(N,X)
    plt.scatter( np.array(centro)[:,0],np.array(centro)[:,1], color='black',marker="*")
    plt.scatter(X[:,0],X[:,1],alpha=0.2)
    plt.xlabel('Skills')
    plt.ylabel('Hardwork')
    plt.show()
```

```python
Final_centro = np.array(centro)                          #Introducing an array for finally determined centeroids
print(Final_centro)
```
